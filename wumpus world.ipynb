```python
#9=stench
#8=glitter
#7=pit
#6=gold
#5=breeze
#-1=wumpus

def learnAgent(world, i, j):
        '''Function for an agent to know what position contains which environment objects'''
        if world[i][j] == 9:
            ai, aj = i, j
            print("Now the agent is at " + str(ai) + "," + str(aj))
            print("Agent came across a stench")
            return ai, aj
        elif world[i][j] == 8:
            ai, aj = i, j
            print("Now the agent is at " + str(ai) + "," + str(aj))
            print("Agent came across a glitter")
            return ai, aj
        elif world[i][j] == 7:
            ai, aj = i, j
            print("Now the agent is at " + str(ai) + "," + str(aj))
            print("Agent came across a pit")
            return -5, -5
        elif world[i][j] == 6:
            ai, aj = i, j
            print("Now the agent is at " + str(ai) + "," + str(aj))
            print("Agent found gold")
            return -4, -4
        elif world[i][j] == 5:
            ai, aj = i, j
            print("Now the agent is at " + str(ai) + "," + str(aj))
            print("Agent feel breeze")
            return ai, aj
        elif world[i][j] == -1:
            ai, aj = i, j
            print("Now the agent is at " + str(ai) + "," + str(aj))
            print("Agent met wumpus")
            return -5, -5
        else:      #if world environment was empty
            ai, aj = i, j
            print("Now the agent is at " + str(ai) + "," + str(aj))
            return ai, aj

def checkInput(ai, aj):
    '''Function for checking input going in forward direction to get gold'''
    if ai == 0 and aj == 0:
        print("Agent can go at " + str(ai+1) + " " + str(aj))      #can move upward
        print("Agent can go at " + str(ai) + " " + str(aj+1))      #can move right
        avi = int(input("Enter input for row => "))
        avj = int(input("Enter input for column => "))
        if avi == ai+1 and avj == aj or avi == ai and avj == aj+1:
            return avi, avj
        else:
            return -5
    elif ai == 3 and aj == 0:
        print("Agent can go at " + str(ai-1) + " " + str(aj))      #can go left
        print("Agent can go at " + str(ai) + " " + str(aj+1))      #can go right
        avi = int(input("Enter input for row => "))
        avj = int(input("Enter input for column => "))
        if avi == ai-1 and avj == aj or avi == ai and avj == aj+1:
            return avi, avj
        else:
            return -5
    elif ai == 3 and aj == 3:
        print("Agent can go at " + str(ai-1) + " " + str(aj))      #can go down
        print("Agent can go at " + str(ai) + " " + str(aj-1))      #can go left
        avi = int(input("Enter input for row => "))
        avj = int(input("Enter input for column => "))
        if avi == ai-1 and avj == aj or avi == ai and avj == aj-1:
            return avi, avj
        else:
            return -5
    elif ai == 0 and aj == 3:
        print("Agent can go at " + str(ai+1) + " " + str(aj))     #can go upward
        print("Agent can go at " + str(ai) + " " + str(aj-1))     #can go left
        avi = int(input("Enter input for row => "))
        avj = int(input("Enter input for column => "))
        if avi == ai+1 and avj == aj or avi == ai and avj == aj-1:
            return avi, avj
        else:
            return -5, -5
    elif ai == 1 and aj == 0 or ai == 2 and aj == 0 or ai == 3 and aj == 0:
        print("Agent can go at " + str(ai+1) + " " + str(aj))     #can go upward
        print("Agent can go at " + str(ai) + " " + str(aj+1))     #can move right
        avi = int(input("\nEnter input for row => "))
        avj = int(input("Enter input for column => "))
        if avi == ai+1 and avj == aj or avi == ai and avj == aj+1:
            return avi, avj
        else:
            return -5, -5
    elif ai == 0 and aj == 3 or ai == 1 and aj == 3 or ai == 2 and aj == 3 or ai == 3 and aj == 3:
        print("Agent can go at " + str(ai+1) + " " + str(aj))     #can go upward
        print("Agent can go at " + str(ai) + " " + str(aj-1))     #can go left
        avi = int(input("Enter input for row => "))
        avj = int(input("Enter input for column => "))
        if avi == ai+1 and avj == aj or avi == ai and avj == aj-1:
            return avi, avj
        else:
            return -5, -5
    elif ai == 3 and aj == 1 or ai == 3 and aj == 2 or ai == 3 and aj == 3:
        print("Agent can go at " + str(ai) + " " + str(aj+1))     #can go right
        print("Agent can go at " + str(ai) + "	" + str(aj-1))     #can go left
        print("Agent can go at " + str(ai-1) + " " + str(aj))     #can move downward
        avi = int(input("Enter input for row => "))
        avj = int(input("Enter input for column => "))
        if avi == ai and avj == aj+1 or avi == ai and avj == aj-1 or avi == ai-1 and avj == aj:
            return avi, avj
        else:
            return -5, -5
    else:
        print("Agent can go at " + str(ai) + " " + str(aj+1))     #can go right
        print("Agent can go at " + str(ai) + " " + str(aj-1))     #can go left
        print("Agent can go at " + str(ai+1) + " " + str(aj))     #can move upward
        avi = int(input("Enter input for row => "))
        avj = int(input("Enter input for column => "))
        if avi == ai and avj == aj+1 or avi == ai and avj == aj-1 or avi == ai+1 and avj == aj:
            return avi, avj
        else:
            return -5, -5

def checkInputReverse(ai, aj):
    '''Function for checking input going in reverse direction to get back to original position'''
    if ai == 0 and aj == 3:
        print("Agent can go at " + str(ai) + " " + str(aj-1))   #can go left
        avi = int(input("Enter input for row => "))
        avj = int(input("Enter input for column => "))
        if avi == ai and avj == aj-1:
            return avi, avj
        else:
            return -5, -5
    elif ai == 0 and aj == 2 or ai == 0 and aj == 1:
        print("Agent can go at " + str(ai) + " " + str(aj+1))   #can go right
        print("Agent can go at " + str(ai) + " " + str(aj-1))   #can go left
        avi = int(input("Enter input for row => "))
        avj = int(input("Enter input for column => "))
        if avi == ai and avj == aj-1 or avi == ai and avj == aj+1:
            return avi, avj
        else:
            return -5, -5
    elif ai == 1 and aj == 0 or ai == 2 and aj == 0:
        print("Agent can go at " + str(ai-1) + " " + str(aj))   #can go downward
        print("Agent can go at " + str(ai) + " " + str(aj+1))   #can move right
        avi = int(input("Enter input for row => "))
        avj = int(input("Enter input for column => "))
        if avi == ai-1 and avj == aj or avi == ai and avj == aj+1:
            return avi, avj
        else:
            return -5, -5
    elif ai == 1 and aj == 3 or ai == 2 and aj == 3:
        print("Agent can go at " + str(ai-1) + " " + str(aj))     #can go downward
        print("Agent can go at " + str(ai) + " " + str(aj-1))     #can go left
        avi = int(input("Enter input for row => "))
        avj = int(input("Enter input for column => "))
        if avi == ai-1 and avj == aj or avi == ai and avj == aj-1:
            return avi, avj
        else:
            return -5, -5
    else:
        print("Agent can go at " + str(ai-1) + " " + str(aj))    #can go downward
        print("Agent can go at " + str(ai) + " " + str(aj-1))    #can go left
        print("Agent can go at " + str(ai) + " " + str(aj+1))    #can go right
        avi = int(input("Enter input for row => "))
        avj = int(input("Enter input for column => "))
        if avi == ai-1 and avj == aj or avi == ai and avj == aj-1 or avi == ai and avj == aj+1:
            return avi, avj
        else:
            return -5, -5

world=[[0,5,7,5],
        [9,0,8,0],
        [-1,6,7,8],
        [9,0,8,7]]   #declaration of a world

ai, aj = 0, 0   #initial agent position
print("Initially agent is at " + str(ai) + "," + str(aj))
print("Agent can go at 	" + str(ai+1) + " " + str(aj))
print("Agent can go at 	" + str(ai) + "	" + str(aj+1))

avi = int(input("Enter input for row => "))
avj = int(input("Enter input for column => "))   #taking row and column values 
if(avi == 1 and avj == 0 or avi == 0 and avj == 1):
    ai, aj = learnAgent(world, avi, avj)   #if input valid calling learn agent function
else:
    print("Not valid move!!")

while(ai >= 0):
    avi, avj = checkInput(ai, aj)
    if(avi != -5 and avj != -5):
        ai, aj = learnAgent(world, avi, avj)
    else:
        print("Not valid move!!")

if(ai == -5):
    print("Game over...Sorry try next time!!!")
else:
    print("You have unlocked next level move back to your initial position") #acquired gold

    ai, aj = 2, 1  #implementation of reverse logic
    while(ai >= 0):
        avi, avj = checkInputReverse(ai, aj)
        if(avi == 0 and avj == 0):
            ai, aj = -4, -4
        elif(avi != -5 and avj != -5):
            ai, aj = learnAgent(world, avi, avj)
        else:
            print("Not valid move!!")

    if(ai == -5):
        print("Really close but unfortunately you failed!!! Try next time")
    else:
        print("Won!!!!!")
```
